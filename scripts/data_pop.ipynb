```python
# import random
# import string
# import mysql.connector
# from datetime import datetime, timedelta

# # Database connection details
# DB_HOST = "172.20.10.10"
# DB_USER = "root"
# DB_PASSWORD = ""
# DB_NAME = "music_platform"

# # Batch config
# insertion_amount = 1
# num_records = 100

# # Helpers
# def random_string(length=10):
#     return ''.join(random.choices(string.ascii_letters, k=length))

# def random_date(start_year=1970, end_year=2025):
#     start = datetime(start_year, 1, 1)
#     end = datetime(end_year, 12, 31)
#     return start + timedelta(days=random.randint(0, (end - start).days))

# def random_country():
#     return random.choice([
#         "Australia", "Cambodia", "Canada", "Vietnam", "France", "Germany",
#         "India", "Indonesia", "Italy", "Japan", "Malaysia", "Singapore",
#         "Spain", "Thailand", "USA"
#     ])

# # Connect
# try:
#     connection = mysql.connector.connect(
#         host=DB_HOST, user=DB_USER, password=DB_PASSWORD, database=DB_NAME
#     )
#     cursor = connection.cursor()
# except mysql.connector.Error as err:
#     print("DB connection error:", err)
#     exit()

# print('Start inserting ...')

# for i in range(insertion_amount):
#     print(f"üîÅ Insertion round {i+1}")

#     subscriber_ids = []
#     artist_ids = []
#     album_ids = []
#     song_ids = []
#     playlist_ids = []
#     subscription_ids = []

#     for _ in range(num_records):
#         # Subscriber
#         username = random_string(8)
#         name = random_string(12)
#         dob = random_date(1970, 2005).strftime('%Y-%m-%d')
#         country = random_country()
#         pwd = random_string(32)
#         cursor.execute("""
#             INSERT INTO subscriber (username, name, dob, country, passwordHash)
#             VALUES (%s, %s, %s, %s, %s)
#         """, (username, name, dob, country, pwd))
#         subscriber_ids.append(cursor.lastrowid)

#         # Artist
#         artist_name = random_string(10)
#         bio = f"Bio of {artist_name}"
#         artist_country = random_country()
#         cursor.execute("""
#             INSERT INTO artist (name, bio, country)
#             VALUES (%s, %s, %s)
#         """, (artist_name, bio, artist_country))
#         artist_ids.append(cursor.lastrowid)

#     # Album and Song
#     for artist_id in artist_ids:
#         album_title = "Album " + random_string(6)
#         album_date = random_date(2000, 2022).strftime('%Y-%m-%d')
#         cursor.execute("""
#             INSERT INTO album (title, release_date, artist_id)
#             VALUES (%s, %s, %s)
#         """, (album_title, album_date, artist_id))
#         album_id = cursor.lastrowid
#         album_ids.append(album_id)

#         for _ in range(2):  # 2 songs per album
#             song_title = "Song " + random_string(5)
#             duration = random.randint(120, 400)
#             release_date = album_date
#             cursor.execute("""
#                 INSERT INTO song (title, duration_in_sec, album_id, release_date)
#                 VALUES (%s, %s, %s, %s)
#             """, (song_title, duration, album_id, release_date))
#             song_id = cursor.lastrowid
#             song_ids.append(song_id)

#             # song_artist
#             cursor.execute("""
#                 INSERT INTO song_artist (artist_id, song_id)
#                 VALUES (%s, %s)
#             """, (artist_id, song_id))

#     # Subscriptions
#     price_lookup = {
#         "Free": 0.00,
#         "Premium": 2.00,
#         "Family": 5.00
#     }

#     for sid in subscriber_ids:
#         plan = random.choice(["Free", "Premium", "Family"])
#         desc = f"{plan} plan with random features"
#         duration = random.choice([30, 60, 180])
#         price = price_lookup[plan]
#         cursor.execute("""
#             INSERT INTO subscription (subscriber_id, plan_name, description, duration_in_days, price)
#             VALUES (%s, %s, %s, %s, %s)
#         """, (sid, plan, desc, duration, price))
#         subscription_ids.append(cursor.lastrowid)

#     # Subscription history
#     for sid, sub_id in zip(subscriber_ids, subscription_ids):
#         start = datetime.now() - timedelta(days=random.randint(0, 100))
#         end = start + timedelta(days=30)
#         cursor.execute("""
#             INSERT INTO subscription_history (subscriber_id, subscription_id, start_date, end_date)
#             VALUES (%s, %s, %s, %s)
#         """, (sid, sub_id, start.strftime('%Y-%m-%d'), end.strftime('%Y-%m-%d')))

#     # Playlists and playlist_song (with duplicate check)
#     for sid in subscriber_ids:
#         title = f"Playlist {random_string(5)}"
#         desc = "desc"
#         created_at = datetime.now().strftime('%Y-%m-%d')
#         is_public = random.choice([True, False])
#         cursor.execute("""
#             INSERT INTO playlist (title, description, created_at, is_public, subscriber_id)
#             VALUES (%s, %s, %s, %s, %s)
#         """, (title, desc, created_at, is_public, sid))
#         pid = cursor.lastrowid
#         playlist_ids.append(pid)

#         songs_added = set()
#         for _ in range(random.randint(1, 3)):
#             song_id = random.choice(song_ids)
#             if song_id in songs_added:
#                 continue
#             songs_added.add(song_id)
#             added_at = datetime.now().strftime('%Y-%m-%d')
#             cursor.execute("""
#                 INSERT INTO playlist_song (playlist_id, song_id, added_at)
#                 VALUES (%s, %s, %s)
#             """, (pid, song_id, added_at))

#     # Play history
#     for sid in subscriber_ids:
#         for _ in range(random.randint(1, 5)):
#             song_id = random.choice(song_ids)
#             played_at = random_date(2022, 2025).strftime('%Y-%m-%d')
#             cursor.execute("""
#                 INSERT INTO play_history (subscriber_id, song_id, played_at)
#                 VALUES (%s, %s, %s)
#             """, (sid, song_id, played_at))

#     # Subscriber follows artist
#     for sid in subscriber_ids:
#         for _ in range(random.randint(1, 3)):
#             artist_id = random.choice(artist_ids)
#             try:
#                 cursor.execute("""
#                     INSERT INTO subscriber_artist (subscriber_id, artist_id)
#                     VALUES (%s, %s)
#                 """, (sid, artist_id))
#             except:
#                 pass  # skip duplicates

#     connection.commit()
#     print(f"‚úÖ Round {i+1}: Inserted {num_records} records (+related data)")

# cursor.close()
# connection.close()
# print("üéâ Finished inserting!")

import random
import string
import mysql.connector
from datetime import datetime, timedelta

# Database connection details
DB_HOST = "localhost"
DB_USER = "root"
DB_PASSWORD = ""
DB_NAME = "music_platform"

# Batch config
insertion_amount = 10
num_records = 100000

# Helpers
def random_string(length=10):
    return ''.join(random.choices(string.ascii_letters, k=length))

def random_date(start_year=1970, end_year=2025):
    start = datetime(start_year, 1, 1)
    end = datetime(end_year, 12, 31)
    return start + timedelta(days=random.randint(0, (end - start).days))

def random_country():
    return random.choice([
        "Australia", "Cambodia", "Canada", "Vietnam", "France", "Germany",
        "India", "Indonesia", "Italy", "Japan", "Malaysia", "Singapore",
        "Spain", "Thailand", "USA"
    ])

# Connect
def connect_db():
    return mysql.connector.connect(
        host=DB_HOST, user=DB_USER, password=DB_PASSWORD, database=DB_NAME
    )

try:
    connection = connect_db()
    cursor = connection.cursor()
except mysql.connector.Error as err:
    print("DB connection error:", err)
    exit()

def safe_execute(query, params):
    global cursor
    global connection
    try:
        if not connection.is_connected():
            print("Reconnecting to the database...")
            connection.reconnect(attempts=3, delay=5)
            cursor = connection.cursor()
        cursor.execute(query, params)
    except mysql.connector.Error as err:
        if err.errno in (2006, 2013):  # MySQL server has gone away / lost connection
            print(f"Lost connection ({err}). Reconnecting and retrying...")
            connection.reconnect(attempts=3, delay=5)
            cursor = connection.cursor()
            cursor.execute(query, params)
        else:
            raise

print('Start inserting ...')

for i in range(insertion_amount):
    print(f"üîÅ Insertion round {i+1}")

    subscriber_ids = []
    artist_ids = []
    album_ids = []
    song_ids = []
    playlist_ids = []
    subscription_ids = []

    # Insert subscribers and artists
    for idx in range(num_records):
        # Subscriber
        username = random_string(10)
        name = random_string(12)
        dob = random_date(1970, 2005).strftime('%Y-%m-%d')
        country = random_country()
        pwd = random_string(32)
        safe_execute("""
            INSERT INTO subscriber (username, name, dob, country, passwordHash)
            VALUES (%s, %s, %s, %s, %s)
        """, (username, name, dob, country, pwd))
        subscriber_ids.append(cursor.lastrowid)

        # Artist
        artist_name = random_string(10)
        bio = f"Bio of {artist_name}"
        artist_country = random_country()
        safe_execute("""
            INSERT INTO artist (name, bio, country)
            VALUES (%s, %s, %s)
        """, (artist_name, bio, artist_country))
        artist_ids.append(cursor.lastrowid)

        # Commit every 20 inserts
        if (idx + 1) % 20 == 0:
            connection.commit()

    # Album and Song
    for artist_id in artist_ids:
        album_title = "Album " + random_string(6)
        album_date = random_date(2000, 2022).strftime('%Y-%m-%d')
        safe_execute("""
            INSERT INTO album (title, release_date, artist_id)
            VALUES (%s, %s, %s)
        """, (album_title, album_date, artist_id))
        album_id = cursor.lastrowid
        album_ids.append(album_id)

        for _ in range(2):  # 2 songs per album
            song_title = "Song " + random_string(5)
            duration = random.randint(120, 400)
            release_date = album_date
            safe_execute("""
                INSERT INTO song (title, duration_in_sec, album_id, release_date)
                VALUES (%s, %s, %s, %s)
            """, (song_title, duration, album_id, release_date))
            song_id = cursor.lastrowid
            song_ids.append(song_id)

            # song_artist
            safe_execute("""
                INSERT INTO song_artist (artist_id, song_id)
                VALUES (%s, %s)
            """, (artist_id, song_id))

    connection.commit()  # Commit albums and songs

    # Subscriptions
    price_lookup = {
        "Free": 0.00,
        "Premium": 2.00,
        "Family": 5.00
    }

    for sid in subscriber_ids:
        plan = random.choice(["Free", "Premium", "Family"])
        desc = f"{plan} plan with random features"
        duration = random.choice([30, 60, 180])
        price = price_lookup[plan]
        safe_execute("""
            INSERT INTO subscription (subscriber_id, plan_name, description, duration_in_days, price)
            VALUES (%s, %s, %s, %s, %s)
        """, (sid, plan, desc, duration, price))
        subscription_ids.append(cursor.lastrowid)

    connection.commit()  # Commit subscriptions

    # Subscription history
    for sid, sub_id in zip(subscriber_ids, subscription_ids):
        start = datetime.now() - timedelta(days=random.randint(0, 100))
        end = start + timedelta(days=30)
        safe_execute("""
            INSERT INTO subscription_history (subscriber_id, subscription_id, start_date, end_date)
            VALUES (%s, %s, %s, %s)
        """, (sid, sub_id, start.strftime('%Y-%m-%d'), end.strftime('%Y-%m-%d')))

    connection.commit()  # Commit subscription history

    # Playlists and playlist_song (with duplicate check)
    for sid in subscriber_ids:
        title = f"Playlist {random_string(5)}"
        desc = "desc"
        created_at = datetime.now().strftime('%Y-%m-%d')
        is_public = random.choice([True, False])
        safe_execute("""
            INSERT INTO playlist (title, description, created_at, is_public, subscriber_id)
            VALUES (%s, %s, %s, %s, %s)
        """, (title, desc, created_at, is_public, sid))
        pid = cursor.lastrowid
        playlist_ids.append(pid)

        songs_added = set()
        for _ in range(random.randint(1, 3)):
            song_id = random.choice(song_ids)
            if song_id in songs_added:
                continue
            songs_added.add(song_id)
            added_at = datetime.now().strftime('%Y-%m-%d')
            safe_execute("""
                INSERT INTO playlist_song (playlist_id, song_id, added_at)
                VALUES (%s, %s, %s)
            """, (pid, song_id, added_at))

    connection.commit()  # Commit playlists

    # Play history
    for sid in subscriber_ids:
        for _ in range(random.randint(1, 5)):
            song_id = random.choice(song_ids)
            played_at = random_date(2022, 2025).strftime('%Y-%m-%d')
            safe_execute("""
                INSERT INTO play_history (subscriber_id, song_id, played_at)
                VALUES (%s, %s, %s)
            """, (sid, song_id, played_at))

    connection.commit()  # Commit play history

    # Subscriber follows artist
    for sid in subscriber_ids:
        for _ in range(random.randint(1, 3)):
            artist_id = random.choice(artist_ids)
            try:
                safe_execute("""
                    INSERT INTO subscriber_artist (subscriber_id, artist_id)
                    VALUES (%s, %s)
                """, (sid, artist_id))
            except mysql.connector.Error:
                # skip duplicates
                pass

    connection.commit()  # Final commit for subscriber_artist

    print(f"‚úÖ Round {i+1}: Inserted {num_records} records (+ related data)")

cursor.close()
connection.close()
print("üéâ Finished inserting!")
```
